```python
import numpy as np
import matplotlib.pyplot as plt

G = 1
```

```python
class Body:
    """a single gravitating body: it has a mass, a position, and a velocity"""

    def __init__(self, mass, position, velocity):
        self.mass = mass
        self.position = np.array(position, dtype=float)
        self.velocity = np.array(velocity, dtype=float)

    def acceleration_from(self, other):
        """acceleration THIS body feels from the gravity of one other body."""
        offset = other.position - self.position  
        distance = np.sqrt(np.sum(offset**2))
        return G * other.mass * offset / distance**3
```

```python
def total_acceleration(bodies, i):
    """net acceleration on bodies[i] from every other body"""
    a = np.zeros_like(bodies[i].position)
    for j in range(len(bodies)):
        if j != i:
            a += bodies[i].acceleration_from(bodies[j])
    return a
```
